```python
import math
import numpy as np
import matplotlib.pyplot as plt

pig=100*np.random.random((2))
print(pig)

V=float(input("input velocity"))
A=float(input("input angle"))
t=np.linspace(0,V*math.sin(math.radians(A))/4.9,1000)

x=V*math.cos(math.radians(A))*t
y=(V*math.sin(math.radians(A))-1/2*9.8*t)*t

plt.plot(x,y,"b-")
plt.plot(pig[0],pig[1],'ro')
plt.xlabel("distance(m)")
plt.ylabel("distance(m)")
plt.show()

px=pig[0]
py=pig[1]
pig_size=5
dx=x-px
dy=y-py
r=(dx**2+dy**2)**(1/2)
a=x[r<pig_size]
print(a)
print(np.shape(a)[0])
if np.shape(a)[0]>0:
    print("you win!")
else:
    print("try again loser")
```
